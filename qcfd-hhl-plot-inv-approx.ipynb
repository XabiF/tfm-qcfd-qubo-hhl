```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(b, x):
    return (1 - (1-x**2)**b) / x
```

```python
kappa = 5
epsilon = 0.1
b = int(np.ceil(kappa**2 * np.log(kappa / epsilon)))

x = np.linspace(-1, 1, 200)
y = abs(f(b, x) - 1/x)

# Plot function
plt.plot(x, y, label=fr"$|f_b(x) - 1/x|$")

# Vertical lines at Â±1/kappa
plt.axvline( 1/kappa, color='red', linestyle='--', label=r"$x=\pm 1/\kappa$")
plt.axvline(-1/kappa, color='red', linestyle='--')

plt.xlabel("x")
plt.ylabel(f"$|f_b(x) - 1/x|$")
plt.grid(True)
print(f"Approximation for $1/x$ ($\\kappa={kappa}$, $\\epsilon={epsilon}$, $b={b}$)")
plt.legend()
plt.show()
```
