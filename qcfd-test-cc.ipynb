```python
import numpy as np
import cfd
import cfdplot
```

```python
def test_psi_bc(sim: cfd.Simulation, x, y):
    xbase = sim.Nx//2
    ybase = sim.Ny//2
    xsize = 3
    ysize = 3

    if (x == 0) or (y == 0) or (x == sim.Nx+1) or (y == sim.Ny+1) or (x == 1) or (y == 1) or (x == sim.Nx) or (y == sim.Ny):
        return test_psi_0(sim, (x-1)*sim.dx, (y-1)*sim.dy)
    
    elif ((x >= xbase) and (x <= xbase+xsize)) and ((y >= ybase) and (y <= ybase+ysize)):
        return 0.0
    
    return None

def test_psi_0(sim: cfd.Simulation, xr, yr):
    umax = 0.5
    return umax*yr
```

```python
sim = cfd.Simulation({
    "log": cfd.LOG_VARIABLES_BCS,
    "icbc": "ldc",
    "Nx": 14, "Ny": 14,
    "Lx": 1.0, "Ly": 1.0,
    "T": 20, "dt": 0.01,
    "nu": 0.1,
    "rho": 1.0,

    # QUBO placeholders
    "chimera-plot": False
})
sim.prepare()
```

```python
# No simulamos, somo hacemos plot de valores iniciales calculados
STEPS_DIR = "test-cc-steps"
sim.plot_steps(STEPS_DIR)
cfdplot.display_step_frame(STEPS_DIR, 0)
```
